Creating Creature class with characteristics, its 3 methods and a test loop

```python
import random
class Creature:

    def __init__(self, name, maxHP=10, a=1, d=5, s=5): #takes name, maxHP attack, defense and speed values with defaults
        self.name = name
        self.HP = maxHP
        self.abilities = {"attack":a, "defense":d, "speed":s}
        self.maxHP = maxHP
        #Flag variables below
        self.modified = False #starts off unmodified
        self.shielded = False #starts off with shield down
        self.alive = True #starts off alive
        return

    def check_life(self):
        if self.HP <= 0: #prevents negative HP by setting min as 0
            self.HP = 0
            print(self.name + " fainted.") #death message
            self.alive = False #set flag as dead
        elif self.HP > self.maxHP: #prevent healing past max HP
            self.HP = self.maxHP
            print(self.name, "is at max health") #maxHP message
        else: 
            print(self.name, "is at", self.HP, "health") #current hp message
        return self.HP

    def attack(self, target, string="attack", roll=None, a=1, b=4):#string is attack type for message, a and b are min and max for damage ints
        print(self.name, string+"s" , target.name) #attack type message and target
        if roll == None: #if no given roll
            roll = random.randint(1,20)
        if roll < (target.abilities["speed"] + target.abilities["defense"]):
            print(string.capitalize(), "missed...") #attack misses if roll less than speed plus defense
            return False #indicates attack missed
        else: #attack succesful
            damage = self.abilities["attack"] + random.randint(a,b) #damage delivered is attack stat + random in given range
            target.HP = target.HP - damage
            print(string.capitalize(), "hits", target.name, "for", damage, "damage!") #danage delivered message
            return True #indicates attack hit
        
    def turn(self, round_num, target):
        if self.attack(target) == True: #calls attack method
            if target.check_life() == 0: #calls check life function after successful attack
                return True #True if target hit and defeated
            else:
                return #nothing if target hit and alive
        else: 
            return False #False if target missed        
 

#short loop for 20 rounds with 2 creatures
creature1 = Creature("Gollum")
creature2 = Creature("Bilbo")

for i in range(1,21):
    print("Round ", i , ":", sep = "")
    if creature1.turn(i,creature2) ==True: #if Gollum beats Bilbo break
        break
    else:
        if creature2.turn(i,creature1) ==True: #if Bilbo beats Gollum Break
            break
```

<br>different defaults
<br>2 unique methods and altered turn method
<br>test loop

```python
class Fighter(Creature): #fighter class inherits from creature class
    def __init__(self, name, MaxHP =50, a=5, d=10, s=3):  #new default values
        Creature.__init__(self, name, MaxHP, a, d, s) #call creature class
        return
        
    def shield_up(self):
        if self.shielded == False: #can't double shield up
            self.abilities["attack"] -= 5
            self.abilities["defense"] += 5
            self.shielded = True #set status as shielded
            print(self.name, "takes a defensive stance.") #status message
        else:
            print(self.name + "'s shield is already up.") #error message, can't call consecutively
        return

    def shield_down(self):
        if self.shielded == True: #same logic as above
            self.abilities["attack"] += 5
            self.abilities["defense"] -= 5
            self.shielded = False
            print(self.name + "'s stance returns to normal.")
        else:
            print(self.name + "'s shield is already down")
        return
        
    def turn(self, round_num, target): #follows new strategy
        round_num %= 4
        if round_num == 1:
            self.attack(target)#call attack method
            self.shield_up()#call shield method
        elif round_num == 2 or round_num == 3: 
            self.attack(target)#call attack method
        else:
            self.shield_down()
            self.attack(target)#call attack method            
        if target.check_life() == 0: 
            return True #indicates they were defeated
        else: 
            return False #indicates still effective in battle

#create short test loop
creature1 = Creature("Gollum")
fighter2 = Fighter("Aragorn")

for i in range(1,21):
    print("Round ", i , ":", sep = "")
    if creature1.turn(i,fighter2) ==True:
        break
    else:
        if fighter2.turn(i,creature1) ==True:
            break
```

```python
class Archer(Creature):
    def __init__(self, name, MaxHP =30, a=7, d=8, s=8): #new defaults
        Creature.__init__(self, name, MaxHP, a, d, s) #inherit Creature class
        return

    def sneak_attack(self, target): #new method
        num1 = random.randint(1,20)
        num2 = random.randint(1,20)
        roll = max(num1, num2) #set roll as max of 2 random ints
        diff = self.abilities["speed"] - target.abilities["speed"] #calculate if faster than target
        if diff > 0:
            roll += (diff)#if faster add the difference to roll
        if self.modified == False: #if first time calling sneak attack
            self.abilities["attack"] += 3 #modify abilities
            self.abilities["defense"] -= 3
            print(self.name + "'s attack rose")
            print(self.name + "'s defense reduced")
            self.modified = True #prevents future modification
        Creature.attack(self, target, "sneak attack",  roll, 1, 8) #call creature attack method with new string, the above roll and the random damage parameters
        return
        
    def attack(self, target): #resets abilities and calls Creature attack method
        self.abilities = {"attack":7, "defense":8, "speed":8} #reset abilities
        Creature.attack(self, target)
        return
        
    def turn(self, round_num, target): #personal round strategy, sneak attacks unless round 1 of 4
        round_num %= 4
        if round_num == 1:
            self.attack(target)
        else:
            self.sneak_attack(target)   
        if target.check_life() == 0:
            return True #indicates target defeated
        else: 
            return False #indicates target is still effective
        
archer1 = Archer("Legolas")
fighter2 = Fighter("Aragorn")
#short test loop
for i in range(1,21):
    print("Round ", i , ":", sep = "")
    if archer1.turn(i,fighter2) ==True:
        break
    else:
        if fighter2.turn(i,archer1) ==True:
            break      



```

```python
class Goblin(Creature):
    
    def __init__(self, name, MaxHP =15, a=4, d=6, s=6): #new default values
        Creature.__init__(self, name, MaxHP, a, d, s) #inherits from creature class
        return
    
class Orc(Creature):
    
    def __init__(self, name, maxHP= 50, a=10, d=6, s=2 ): #new deaults
        Creature.__init__(self, name, maxHP, a, d, s) #inherits from creature class
        self.unmodified = False #stops orc from cooling down each attack after original ability change
        return
    
    def heavy_attack(self, target): #new unique method
        if self.modified == False: #check if abilities have been changed
            self.abilities["attack"] += 5
            self.abilities["defense"] -= 3
            print(self.name, "is in a rage.")
            self.modified = True
        Creature.attack(self, target, "heavy attack")#call original attack method from creature class
        return
    
    def attack(self, target):
        if self.modified == True and self.unmodified == False:
            self.abilities["attack"] -= 5
            self.abilities["defense"] += 3
            print(self.name, "cooled down.")
            self.unmodified = True #stops orc from cooling down each attack after original ability change
        Creature.attack(self, target) #call original attack method
        return
    
    def turn(self, round_num, target):
        round_num %= 4
        if round_num == 0: #round 4 heavy attack
            self.heavy_attack(target)
        else: #rounds 1-3 just attack mehod
            self.attack(target)
        if target.check_life() == 0:
            return True #indicate target defeated
        else: 
            return False #indicate target survived
        
goblin1 = Goblin("Gobbo")
orc2 = Orc("Orco")
#short test loop
for i in range(1,21):
    print("Round ", i , ":", sep = "")
    if goblin1.turn(i,orc2) ==True:
        break
    else:
        if orc2.turn(i,goblin1) ==True:
            break  
```

```python
class Wizard(Creature):
    def __init__(self, name, maxHP=20):
        Creature.__init__(self, name, maxHP, 3, 5, 5)
        self.abilities["arcana"] = 10 
        self.mana = 100
        self.arcDam = self.abilities["arcana"]//2 #useful variable, just half of arcana value
        return 
    
    def mana_check(self , min_mana =0): #default min as 0
        if self.mana >= 100:
            self.mana = 100
            print("Mana is full")
            return True
        elif self.mana < min_mana: #check against argument
            print("Not enough mana")
            return False
        else:
            print("Current mana is:", self.mana)
            return
    
    def attack(self, target):
        self.mana += 20 #regains 20 mana, checks levels, and then attacks target
        print("Mana: +20!")
        self.mana_check()
        if Creature.attack(self, target)==True:
            target.check_life()#check if fainted
        return
    
    def recharge(self):
        self.mana += 30 #regains 30 mana
        print(self.name,"is channelling magical energy...\nMana: +30!")
        self.mana_check() #check mana status
        return
    
    def fire_bolt(self, target):
        roll = random.randint(1,20)
        roll += self.arcDam #adds half arcana damage to roll rounded down
        x, self.abilities["attack"] = self.abilities["attack"], 0  #remove attack skills influence on Creature method attack damage     
        if Creature.attack(self, target, "firebolt", roll, 1, self.abilities["arcana"]) == True: #call creature attack, new string, above roll, damage range 1-arcana value
            #if attack hits add 10 mana
            self.mana += 10
            print("Mana: +10!")
            self.mana_check()#check not above max
            self.abilities["attack"] = x #return attack ability back to normal
            target.check_life()#check if fainted
        return
    
    def heal(self, ally):
        if self.mana_check(20) != False: #check if have enough mana
            self.mana -= 20 #cost of spell
            print("Mana:", -20 )
            healVal = random.randint(0, 8) + self.arcDam #heal value specified in question
            ally.HP += healVal
            print(self.name, "healed", ally.name, "for:", healVal, "HP.")
            ally.check_life()
        return
    
    
    def mass_heal(self, allies):
        if self.mana_check(30) != False: #check if has min mana           
            self.mana -= 30 #cost of spell
            print("Mana:", -30 )
            healVal = random.randint(0,10) + self.abilities["arcana"] #healval same as before but 10 not 8
            self.HP += healVal #heal self
            print(self.name, "healed themself for:", healVal, "HP.")
            self.check_life()
            for ally in allies: #heal allies
                ally.HP += healVal
                print(self.name, "healed", ally.name, "for:", healVal, "HP.")
                ally.check_life()
        return
        
    def fire_storm(self, enemies):
        if self.mana_check(50) != False:
            self.mana -= 50 #cost of spell
            print("Mana:", -50 )            
            roll1 = random.randint(1,20) + self.abilities["speed"] #adds speed to random int
            if roll1 >= self.abilities["arcana"]: #if result greater or equal to arcana deliver half damage
                a = 0.5 #half
            else:#take full damage
                a = 1 #full
            for enemy in enemies:
                dam = a*(random.randint(5,20)+ self.abilities["arcana"]) #modifier x 5-20 + arcana value
                enemy.HP -= dam
                print("Fire Storm deals", dam, "fire damage to", enemy.name +"!")
                enemy.check_life()#check if fainted
        return 

creature1 = Creature("Gollum")
creature2 = Creature("Bilbo")
fighter2 = Fighter("Aragorn")
archer1 = Archer("Legolas")
goblin1 = Goblin("Gobbo")
orc2 = Orc("Orco")
goblin2 = GoblinKing("Gobbo")
orc3 = OrcGeneral("Orco")

wiz1 = Wizard("Gandalf")
allies = [creature2, fighter2, archer1]
enemies = [creature1, goblin1, orc2, goblin2, orc3]
wiz1.attack(fighter2)
print()
wiz1.fire_bolt(creature2)
print()
wiz1.heal(creature2)
print()
wiz1.mass_heal(allies)
print()
wiz1.fire_storm(enemies)
print()
wiz1.recharge()
```

```python
class Battle:
    
    def __init__(self):
        self.GobK = GoblinKing("The Goblin King")
        self.OrcG = OrcGeneral("The Orc General")
        self.Gob = Goblin("Goblin")
        self.Orc = Orc("Orc")
        self.enemies = [self.GobK, self.OrcG, self.Gob, self.Orc]
        
        self.Fig = Fighter("The Fighter")
        self.Arc = Archer("The Archer")
        self.allies = [self.Fig, self.Arc]
        
        self.wiz1 = Wizard("The Wizard")
        return
    
    #new method returns random target from target list
    def auto_select(self, target_list): #requires dead enemies to have been removed by list
        if len(target_list) == 0: #if list is empty return None
            return None
        else:
            t = random.randint(0, len(target_list)-1) #choose random target from list
            return target_list[t] #return target
        
    def select_target(self, target_list): #method to select player's target
        print("Select target:")
        for i in range(0, len(target_list)): #display targets and stats
            print(str(i+1)+":", target_list[i].name +", HP:", str(target_list[i].HP) +"/"+ str(target_list[i].maxHP))
        inp = int(input("Enter choice: ")) #ask for choise
        while (0 < inp <= len(target_list)) == False : #if input not any of the targets ask for new input
            print("That choice is not valid, choose again")
            inp = int(input("Enter choice: "))
        inp -= 1 #used to align with indexes
        print("You selected", target_list[inp].name)
        return target_list[inp] #return target

    #new simulates battle method without player   
    

    def start(self):
        self.spacer = "=======================================================" #self.spacer for presentation later
        merged = self.enemies + self.allies #combine all creatures to a list
        merged.append(self.wiz1)#add wiz to list for ordering
        maxspeed = 0 #set max speed var to 0
        for person in merged: #find max speed of characters
            if person.abilities["speed"] > maxspeed:
                maxspeed = person.abilities["speed"]
                
        fast_to_slow = []
        for i in range(0, maxspeed+1): #use max speed for range
            for person in merged: #sort characters fastest to slowest
                if person.abilities["speed"] == i:
                    fast_to_slow.insert(0, person)
        
        round_num = 1 #set for first round
        print("THE BATTLE BEGINS\n" + self.spacer)
        while self.enemies !=[] and self.allies != [] and self.wiz1.alive ==True and round_num<1000: #while no victory conditions met and preventing infinite loop 
            print("Round:", round_num, "\n" + self.spacer) #print round number   
            for char in fast_to_slow: #play through character list
                if char.alive == False: #dead character doesnt get a turn
                    continue
                if char in self.enemies: #enemies choose random target
                    self.temp_list = self.allies.copy()
                    self.temp_list.append(self.wiz1) #add wizard to list of targets for the enemies
                    targ = self.auto_select(self.temp_list)
                    char.turn(round_num, targ) #uses target selected for argument for character's turn method
                elif char in self.allies: #allies choose random enemy target
                    targ = self.auto_select(self.enemies)
                    char.turn(round_num, targ) #uses target selected for argument for character's turn method
                else: 
                    if self.player_turn() == False:
                        return 
                        
                print(self.spacer)
                for enemy in self.enemies: #removes dead enemies from battle
                    if enemy.alive == False:
                        self.enemies.remove(enemy)
                for ally in self.allies:
                    if ally.alive == False:
                        self.allies.remove(ally)
                                           
                  
                    #if wiz dies the while loop ends and skips to end message below
               
                if self.enemies == []: #checks victory conditions and terminates if any tru
                    print("The battle is over")
                    print("The Allies won")
                    break
                elif self.allies == []:
                    print("The battle is over")
                    print("The Enemies won")
                    break
                elif self.wiz1.alive == False:
                    print("The battle is over")
                    print("You died")
                    print("The Enemies won")
                    break
            print("End of round", round_num)
            print(self.spacer)
            round_num += 1 #start next round
            
        return
    
    def player_turn(self): #called on wizards turn
        W = self.wiz1 #set shortened variable
        #print info ofplayer stats and allies
        print("Player:", W.name, "HP:"+ str(W.HP) + "/" + str(W.maxHP), "Mana:", str(W.mana) +"/100") 
        print("Allies:")
        for i in self.allies:
            print(i.name, "HP:" + str(i.HP)+"/"+str(i.maxHP))
        print(self.spacer)
        #print available options/moves for player
        print("Actions. F: Attack R: Recharge Mana\nSpells. 1: Heal 2: Firebolt 3: Mass Heal 4: Fire Storm\nTo Quit game type: Quit")
        print(self.spacer)
        
        act = input("Enter action: ")#ask to select player move
        while act not in  "1234FfRr" and act != "Quit" and act != "quit": #ensure valid input type
            print("Invalid Input, Try Again")
            act = input("Enter action: ")

        if act == "1": #if heal selected, show targets available to heal
            print("Select target")
            available = [0] #list of available targets
            print("0:", W.name+", HP:"+str(W.HP)+"/"+str(W.maxHP))
            for i in range(0,len(self.allies)):
                print(str(i+1)+":", self.allies[i].name + ", HP:"+str(self.allies[i].HP)+"/"+str(self.allies[i].maxHP))
                available.append(i+1)
            select = int(input("Enter selection: ")) #ask for choice of targets
            while select not in available: #check for valid input
                select = int(input("Enter a VALID selection: ")) #ask for choice of targets
            print(self.spacer)
            if select == 0:
                W.heal(W)
            else: 
                W.heal(self.allies[select-1]) #heal an ally


        elif act in "2Ff":
            print("Select target")
            available = [] #list of available targets
            for i in range(0,len(self.enemies)):
                print(str(i+1)+":", self.enemies[i].name + ", HP:"+str(self.enemies[i].HP)+"/"+str(self.enemies[i].maxHP))
                available.append(i+1)
            select = int(input("Enter selection: ")) #ask for choice of targets
            while select not in available: #check for valid input
                select = int(input("Enter a VALID selection: ")) #ask for choice of targets
            if act == "2":
                W.fire_bolt(self.enemies[select-1])
                print(self.spacer)
            elif act in "Ff":
                W.attack(self.enemies[select-1])
                print(self.spacer)
        
        elif act in "Rr":
            W.recharge()
            print(self.spacer)
        elif act == "3":
            W.mass_heal(self.allies)
            print(self.spacer)
        elif act == "4":
            W.fire_storm(self.enemies)
            print(self.spacer)
        else:
            print("You have chosen to quit")
            return False
        
        for enemy in self.enemies: #removes dead enemies from battle
            if enemy.alive == False:
                self.enemies.remove(enemy)
        
        return
            
batt = Battle()
batt.start()
exit()
```
